```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
def initial_cleaning(json_file):
    df = pd.read_json(json_file, orient='index')
    df = pd.DataFrame(df.cards[1])
    df = df[df['promoTypes'].isnull()]
    df = df.loc[df['securityStamp'] != 'arena']
    df = df.loc[~df['supertypes'].map(lambda x: bool('Basic' in x))]
    df = df.loc[:, ['name', 'colors', 'manaValue', 'manaCost', 'power', 'toughness', 
                    'loyalty', 'text', 'rarity', 'keywords', 'supertypes', 'types']]
    #df.set_index('name', inplace=True)
    to_unlist = ['colors', 'types', 'supertypes', 'keywords']
    df['keywords'].fillna('', inplace=True)
    for column in to_unlist:
        df[column] = [', '.join(map(str, l)) for l in df[column]]
    to_convert = ['power', 'toughness', 'loyalty']
    for column in to_convert:
        df[column] = pd.to_numeric(df[column], errors='coerce')
    return df
```

```python
files = ['AFR', 'ELD', 'IKO', 'KHM', 'MID', 'NEO', 'SNC', 'STX', 'THB', 'VOW', 'ZNR']
```

```python
dfs = []
```

```python
for f in files:
    dfs.append(initial_cleaning('datasets/'+f+'.json'))
```

```python
def jitter_arr(arr):
    stdev = .02 * (max(arr) - min(arr))
    return arr + np.random.randn(len(arr)) * stdev
```

```python
plt.figure(figsize=(12,8))
plt.scatter(x = afr_creatures['manaValue'], y = jitter_arr(afr_creatures['power']), color = 'red', alpha=.5, label='Power')
plt.scatter(x = afr_creatures['manaValue'], y = jitter_arr(afr_creatures['toughness']), color = 'green', alpha =.5, label = 'Toughness')
plt.legend(fontsize = 12)
plt.show()
```

```python
minval = 0
maxval = 10
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim((minval, maxval))
ax.set_ylim((minval, maxval))
xs = afr_creatures['power']
ys = afr_creatures['toughness']
zs = afr_creatures['manaValue']
ax.scatter(jitter_arr(xs), jitter_arr(ys), jitter_arr(zs), c=('r'))

ax.set_xlabel('Power')
ax.set_ylabel('Toughness')
ax.set_zlabel('Mana Value')
ax.set_title('Power & Toughness of Creatures vs. Mana Value for "Adventures in the Forgotten Realms"', pad=0)

plt.show()
```
